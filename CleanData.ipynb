```python
# Importing required libraries
import pandas as pd
from pathlib import Path

# Setting variables to iterate through files names in for loop
years = ["2018", "2019", "2020", "2021", "2022"]
years_2018_2020 = ["2018", "2019", "2020"]
months = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
concat_df = pd.DataFrame()
```

## Combining like files Jan 2018 - Jan 2021

```python
# File format was the same from 2018-01-01 through 2021-01-31 so this cell will combine those files into single dataframe

# Iterate through the full years that are the same at the beginning, concatonating along the way
for year in years_2018_2020:
    for month in months:
        df = pd.read_csv("raw-data/JC-" + year + month + "-citibike-tripdata.csv")
        print(f"{year}{month} length: {len(df)}")
        concat_df = pd.concat([concat_df, df])
    print(f"{year} concat df length: {len(concat_df)}")

# Add the first month in 2021 since it is the same
df_202101 = pd.read_csv("raw-data/JC-202101-citibike-tripdata.csv")
print(f"df202101 length: {len(df_202101)}")
concat_df_201801_202101 = pd.concat([concat_df, df_202101])
print(f"concat df length: {len(concat_df_201801_202101)}")

# concat_df_201801_202101.to_csv("clean-data/201801-202101-citibike-tripdata.csv")

concat_df_201801_202101.head()     
```

## Combining like files Jan 2018 - Jan 2021

Manually had to update 1 file name (JC-202207-citbike-tripdata.csv) for consistent naming to JC-20YYMM-citibike-tripdata.csv

```python
# File format was the same from 2021-02-01 through 2022-12-31 so this cell will combine those files into single dataframe

months_no1 = ["02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
concat_df2 = pd.DataFrame()

# Iterate through the remaining months of 2021, concatonating along the way
for month in months_no1:
    df = pd.read_csv("raw-data/JC-2021" + month + "-citibike-tripdata.csv")
    print(f"2021{month} length: {len(df)}")
    concat_df2 = pd.concat([concat_df2, df])
print(f"2021 concat df length: {len(concat_df2)}")

# Iterate through the all months of 2022, concatonating along the way
for month in months:
    df = pd.read_csv("raw-data/JC-2022" + month + "-citibike-tripdata.csv")
    print(f"2022{month} length: {len(df)}")
    concat_df2 = pd.concat([concat_df2, df])
print(f"2022 concat df length: {len(concat_df2)}")

concat_df_202102_202212 = concat_df2

# concat_df_202102_202212.to_csv("clean-data/202102-202212-citibike-tripdata.csv")

concat_df_202102_202212.head()  
```

## Updating headers so that the two formats can be merged into one dataframe

```python
# Update older headers to the new name with exception of usertype column
concat_df_201801_202101.rename(columns={"starttime":"started_at", "stoptime":"ended_at", 
                                        "start station id":"start_station_id", "start station name":"start_station_name", 
                                        "start station latitude":"start_lat", "start station longitude":"start_lng", 
                                        "end station id":"end_station_id","end station name":"end_station_name", 
                                        "end station latitude":"end_lat", "end station longitude":"end_lng"}, inplace=True)
concat_df_201801_202101.head()
```

```python
# Update new header member_casual to align with the older name, usertype
concat_df_202102_202212.rename(columns={"member_casual":"usertype"}, inplace=True)
concat_df_202102_202212.head()
```

## Merge the two dataramges leaving the different columns with null values

```python
# Now that header names are aligned, combine all into one df
concat_df_2018_2022 = pd.concat([concat_df_201801_202101, concat_df_202102_202212])
concat_df_2018_2022
```

# Merge lat / long differences to average lat / long by station
This avoids having many ride counts of 1 at the same station name due to very minor differences in lat / long numbers

## Update start station lat / long values to the average based on station id

```python
# grab the list of latitude and longitude values
station_start_lats = concat_df_2018_2022.groupby(["start_station_id"])["start_lat"].mean()
station_start_lngs = concat_df_2018_2022.groupby(["start_station_id"])["start_lng"].mean()

# iterate through the list of stations modifying both lat and long
for station_id in station_start_lats.index:
    concat_df_2018_2022.loc[concat_df_2018_2022.start_station_id == station_id, "start_lat"] = station_start_lats[station_id]
    concat_df_2018_2022.loc[concat_df_2018_2022.start_station_id == station_id, "start_lng"] = station_start_lngs[station_id]

concat_df_2018_2022
```

## Update end station lat / long values to the average based on station id

```python
# grab the list of latitude and longitude values
station_end_lats = concat_df_2018_2022.groupby(["start_station_id"])["end_lat"].mean()
station_end_lngs = concat_df_2018_2022.groupby(["start_station_id"])["end_lng"].mean()

# iterate through the list of stations modifying both lat and long
for station_id in station_end_lats.index:
    concat_df_2018_2022.loc[concat_df_2018_2022.end_station_id == station_id, "end_lat"] = station_end_lats[station_id]
    concat_df_2018_2022.loc[concat_df_2018_2022.end_station_id == station_id, "end_lng"] = station_end_lngs[station_id]

concat_df_2018_2022
```

## Export the dataframe

```python
concat_df_2018_2022.to_csv("clean-data/2018-2022-citibike-tripdata-JC.csv")
```
